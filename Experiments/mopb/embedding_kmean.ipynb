```python
import numpy as np
datapath = "../../Preprocessing/Encoded_Data/mopb_embeddings.npz"
data = np.load(datapath)
embeddings = data["embeddings"]
labels = data["labels"]
```

```python
embeddings = [emb[0] for emb in embeddings]
```

```python
from sklearn.cluster import MiniBatchKMeans
from sklearn.metrics import v_measure_score

kmeans = MiniBatchKMeans(n_clusters=13, random_state=42)
pred_labels = kmeans.fit_predict(embeddings)

v_measure = v_measure_score(labels, pred_labels)
print(f"V-Measure Score: {v_measure:.4f}")
print(f"Number of clusters found: {len(set(pred_labels)) - (1 if -1 in pred_labels else 0)}")
print(f"Number of noise points: {list(pred_labels).count(-1)}")
```
