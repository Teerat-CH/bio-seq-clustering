```python
import matplotlib.font_manager as fm
import requests

url = "https://github.com/google/fonts/raw/refs/heads/main/ofl/lora/Lora%5Bwght%5D.ttf"
response = requests.get(url, allow_redirects=True)
font_path = "/tmp/Lora-Regular.ttf"

with open(font_path, "wb") as f:
    f.write(response.content)

fm.fontManager.addfont(font_path)

import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Lora'
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import umap

def plot_umap(embeddings, labels, title="UMAP", point_size=5):
    reducer = umap.UMAP(random_state=42)
    emb_2d = reducer.fit_transform(embeddings)

    unique_labels = np.unique(labels)
    cmap = plt.get_cmap("tab20", len(unique_labels))
    color_map = {label: cmap(i) for i, label in enumerate(unique_labels)}
    colors = [color_map[l] for l in labels]

    fig, ax = plt.subplots(figsize=(7, 5), facecolor="white")
    ax.set_facecolor("white")

    ax.scatter(*emb_2d.T, c=colors, s=point_size, alpha=0.7, zorder=2)

    patches = [mpatches.Patch(color=color_map[l], label=l) for l in unique_labels]
    ax.legend(
        handles=patches,
        bbox_to_anchor=(1.05, 1),
        loc="upper left",
        fontsize=8,
        frameon=True,
        edgecolor="#cccccc",
        facecolor="white"
    )

    ax.set_title(title, fontsize=14, fontweight="bold", fontfamily="Lora", pad=12)
    ax.set_xlabel("UMAP 1", fontsize=11, fontfamily="Lora")
    ax.set_ylabel("UMAP 2", fontsize=11, fontfamily="Lora")

    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_color("#cccccc")
    ax.spines["bottom"].set_color("#cccccc")

    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
datapath = "./Encoded_Data/mopb_embeddings.npz"
data = np.load(datapath)
embeddings = data["embeddings"]
labels = data["labels"]
mid_layer  = embeddings[:, 0, :]
last_layer = embeddings[:, 1, :]
embeddings_pooled = embeddings.mean(axis=1)
plot_umap(mid_layer, labels, title="UMAP plot of MopB Foundation Model Embeddings - Mid Layer")
plot_umap(last_layer, labels, title="UMAP plot of MopB Foundation Model Embeddings - Last Layer")
plot_umap(embeddings_pooled, labels, title="UMAP plot of MopB Foundation Model Embeddings - Pooled")
```
